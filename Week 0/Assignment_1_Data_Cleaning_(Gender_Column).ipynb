```python
FILE_PATH ='/content/drive/MyDrive/CARISURG/WEEK_0/EmergencyTriageDataset_Reduced_Dirty.csv'

df_raw = pd.read_csv(FILE_PATH)
print(f"Dataset successfully loaded: {df_raw.shape[0]} rows x {df_raw.shape[1]} columns")
print(f"Columns Headings: {list(df_raw.columns)}")
```

```python
# Cleaning the Gender Column
print("Unique Values in Gender:")
print(df_raw['Gender'].unique())
print(f"\nTotal number of different values: {df_raw['Gender'].nunique()}")
```

```python
# Counting the total gender value
print("Gender Value total:")
print(df_raw['Gender'].value_counts())
```

```python
# Cleaning Gender

gender_map = {
    'Female': 'F',
    'FEMALE': 'F',
    '0':      'F',
    'Male':   'M',
    'MALE':   'M',
    '1':      'M'
}
df_raw['Gender_Clean'] = df_raw['Gender'].astype(str).map(gender_map)
```

```python
#Verifying
print("After cleaning:")
print(df_raw['Gender_Clean'].value_counts(dropna=False))
print(f"\nAny NaN values? {df_raw['Gender_Clean'].isnull().sum()}")
```

```python
#Replacing Original Column

df_raw = df_raw.drop(columns=['Gender'])
df_raw = df_raw.rename(columns={'Gender_Clean': 'Gender'})

print("'Gender' Column after cleaning:")
print(df_raw['Gender'].value_counts())
print(f"Data type: {df_raw['Gender'].dtype}")

df_raw.head()
```
